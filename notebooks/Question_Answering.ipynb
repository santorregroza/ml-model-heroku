# Question Answering

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

```python
model_name = "deepset/roberta-base-squad2"
```

```python
nlp = pipeline("question-answering", model=model_name, tokenizer=model_name)
```

```python
context = "We are enjoying the summer and preparing the PyCon workshop that will be in October, we are in Spain and the workshop is online"
questions = ["What are we doing?", "What are we enjoying?", "What are we preparing?", "When is the PyCon?", "Where are we?", "Where is the workshop?"]
```

```python
for question in questions:

  input = {
              "question": question,
              "context": context
          }
          
  res = nlp(input)
  print(res)
```
